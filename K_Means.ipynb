```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
#Iris Dataset
iris = datasets.load_iris()
X = iris.data

# Coletando as distorções em uma lista
distortions = []
K = range(1,10)
for k in K:
  model = KMeans(n_clusters=k)
  model.fit(X)
  distortions.append(model.inertia_)

# Plotando as distorções
#plt.figure(figsize=(12,6))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distorção')
plt.title('O método do cotovelo (Elbow Method) mostrando o número ótimo de clusters')
plt.show()
```

```python
#importando as bibliotecas
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
# Gerando um dataset aleatório
A = np.random.rand(50,2)
B = 2 + np.random.rand(50,2)
X = np.concatenate((A,B))
```

```python
# Treinando o modelo e fazendo as predições
KMean= KMeans(n_clusters=2, random_state=123)
KMean.fit(X)
label=KMean.predict(X)
```

```python
# avaliando o modelo com o método da silhueta
print(f'Silhouette Score(n=2): {silhouette_score(X, label)}')
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
%matplotlib inline
from sklearn import datasets
from sklearn.metrics import silhouette_score
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
%matplotlib inline
from sklearn import datasets
from sklearn.metrics import silhouette_score
```

```python
#Wine Dataset
wines = datasets.load_wine()
X = wines.data
```

```python
alcool      = X[:, 0]
malico      = X[:, 1]
magnesio    = X[:, 4]
flavonoides = X[:, 6]
```

```python
# Coletando as distorções em uma lista
distortions = []
K = range(1,5)
for k in K:
  model = KMeans(n_clusters=k)
  model.fit(X)
  distortions.append(model.inertia_)

# Plotando as distorções
#plt.figure(figsize=(12,6))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distorção')
plt.title('O método do cotovelo (Elbow Method) mostrando o número ótimo de clusters')
plt.show()
```

```python
km = KMeans(n_clusters=4)
```

```python
km.fit(X)
```

```python
km.predict(X)
labels = km.labels_
```

```python
labels
```

```python
fig = plt.figure(1, figsize=(7,7))
ax = Axes3D(fig, rect=[0, 0, 0.95, 1], elev=48, azim=134)
ax.scatter(alcool, malico, magnesio, flavonoides,
c=labels.astype(np.float), edgecolor="k", s=50)
ax.set_xlabel("Width")
ax.set_ylabel("Height")
ax.set_zlabel("depth")
plt.title("K Means", fontsize=14)
```

```python
# avaliando o modelo com o método da silhueta
print(f'Silhouette Score: {silhouette_score(X, labels)}')
```
